      <li><span class="badge">corr_matrix = numeric_df.corr()</span> — Pairwise correlations (default: Pearson, range −1 to +1).</li>
      <li><span class="badge">sns.heatmap(corr_matrix, ...)</span> — Renders the correlation matrix as a heatmap.</li>
      <li><span class="badge">cbar=True</span>, <span class="badge">cbar_kws={'orientation':'horizontal'}</span> — Shows a horizontal color bar.</li>

```python
# --------------------------------------------------
# STEP 3: Select only numeric columns
# --------------------------------------------------
# .select_dtypes(include=["number"]) drops any column that isn’t a numeric dtype
numeric_df = df.select_dtypes(include=["number"])

# --------------------------------------------------
# STEP 4: Compute the correlation matrix on numeric columns
# --------------------------------------------------
corr_matrix = numeric_df.corr()

# --------------------------------------------------
# STEP 5: Plot the heatmap
# --------------------------------------------------
plt.figure(figsize=(12, 12))

# Create a diverging color palette (red for negative, blue for positive)
cmap = sns.diverging_palette(10, 220, as_cmap=True)

sns.heatmap(
    corr_matrix,
    annot=True,             # show correlation coefficients in each cell
    fmt=".2f",              # 2 decimal places
    cmap=cmap,              # use the custom diverging palette
    center=0,               # center the colormap at 0
    linewidths=0.1,         # width of grid lines between cells
    linecolor="white",      # grid line color
    square=True,            # make each cell square-shaped
    cbar=True,              # display a color bar
    cbar_kws={"orientation": "horizontal"}  # put the color bar at the bottom
)

plt.title("Correlation Matrix Heatmap (Numeric Columns)")
plt.xticks(rotation=45, ha="right")  # rotate x‐labels for readability
plt.yticks(rotation=0)                # keep y‐labels horizontal
plt.tight_layout()
plt.show()
```
